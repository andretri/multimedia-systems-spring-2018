```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def ConvertGrayScale(rgbImage):
    return np.dot(rgbImage[:,:,:3], [0.2989, 0.587, 0.114])
```

```python
def QuantizeImage(img, Q=2):
    return np.around(img*0.9, Q)
```

```python
img = plt.imread('resources/lenna.png')
```

```python
plt.imshow(img)
```

```python
img2 = ConvertGrayScale(img)
plt.imshow(img2, cmap='gray')
```

```python
img3 = QuantizeImage(img2, Q=2)
plt.imshow(img3, cmap='gray')
```
